```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('penguins')
df
```

* displot

```python
sns.displot(df['body_mass_g'])
plt.show()
```

```python
sns.set_theme(style='darkgrid')
sns.displot(df['body_mass_g'])
plt.show()
```

```python
sns.set_theme(style="white", palette="pastel")

sns.displot(df, x="body_mass_g", 
            col="species", 
            row="sex", 
            facet_kws=dict(margin_titles=True))
plt.show()
```

* jointplot 

```python
sns.jointplot(data = df, x = "bill_length_mm", y = "bill_depth_mm")
plt.show()
```

```python
sns.jointplot(data = df, x = "bill_length_mm", y = "bill_depth_mm", kind="kde")
plt.show()
```

```python
sns.jointplot(data = df, x = "bill_length_mm", y = "bill_depth_mm", hue="species")
plt.show()
```

* pairplot

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df, hue="species", height=1.5, palette='plasma')
plt.show()
```

### Ejercicio

* Usar otro dataset de seaborn y hacer displot, jointplot y pairplot.

```python
# Obtención de los nombres de los datasets disponibles
sns.get_dataset_names()
```

```python
# Carga del dataset de interés
df_s = sns.load_dataset('iris')
df_s
```

* displot

```python
sns.displot(df_s['species'])
plt.show()
```

```python
sns.set_theme(style='whitegrid')
sns.displot(df_s, x='petal_length', col='species', facet_kws=dict(margin_titles=True))
plt.show()
```

```python
sns.set_theme(style='whitegrid')
sns.displot(df_s, x='petal_length', y='petal_width', col='species', facet_kws=dict(margin_titles=True))
plt.show()
```

```python
sns.set_theme(style='whitegrid')
sns.displot(df_s, x='sepal_length', col='species', facet_kws=dict(margin_titles=True))
plt.show()
```

```python
sns.set_theme(style='whitegrid')
sns.displot(df_s, x='sepal_length', y='sepal_width', col='species', facet_kws=dict(margin_titles=True))
plt.show()
```

```python
sns.displot(df_s, x='petal_length', hue='species')
plt.show()
```

* jointplot

```python
sns.jointplot(data = df_s, x = "petal_length", y = "petal_width", hue='species')
plt.show()
```

```python
sns.jointplot(data = df_s, x = "petal_length", y = "petal_width", hue='species', kind='kde')
plt.show()
```

* pairplot

```python
sns.pairplot(df_s, hue="species", height=2, palette='rainbow')
plt.show()
```
