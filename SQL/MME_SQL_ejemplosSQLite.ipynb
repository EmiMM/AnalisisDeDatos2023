```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sqlite3

database = 'database.sqlite'
```

```python
# Acceso a la base de datos

conn = sqlite3.connect(database)

tables = pd.read_sql("""SELECT * 
                        FROM sqlite_master
                        WHERE type='table';""", conn)

tables
```

```python
# Tabla de países
countries = pd.read_sql("""SELECT *
                            FROM Country;""", conn)
countries
```

```python
# Lista de países ordenada de forma descendente alfabeticamente
countries = pd.read_sql("""SELECT id, name
                            FROM Country
                            ORDER BY name DESC;""", conn)
countries
```

```python
# Tabla de atributos de jugadores
players = pd.read_sql("""SELECT *
                         FROM Player_Attributes;""", conn)
players
```

```python
pd.read_sql(""" SELECT * FROM player LIMIT 5;""", conn)
```

* Order by

```python
# Tabla de equipos ordenada de forma ascendente por team_fifa_api_id
teams = pd.read_sql("""SELECT *
                        FROM Team
                        ORDER BY team_fifa_api_id ASC;""", conn)
teams
```

```python
pd.read_sql(""" SELECT player_name, height
                FROM Player 
                ORDER BY height DESC
                LIMIT 5;""",conn)
```

* Where

```python
pd.read_sql("""SELECT player_name, weight, height
                FROM Player
                WHERE height <= 180
                LIMIT 7;""",conn)
```

```python
pd.read_sql("""SELECT player_name, weight, height
                FROM Player
                WHERE height <= 180 AND weight < 150 
                LIMIT 7;""",conn)
```

# Ejercicio

* ¿Cuántos equipos diferentes hay?
* ¿Cuántos jugadores con nombre Michael cumplen años en 1992?

```python
# Número de equipos diferentes
pd.read_sql("""SELECT COUNT(DISTINCT(team_long_name)) AS Equipos_distintos
                FROM Team;""",conn)
```

```python
# Número de jugadores con nombre Michael que cumplen años en 1992
pd.read_sql("""SELECT COUNT(*)
                FROM Player
                WHERE player_name LIKE 'Michael%' AND birthday LIKE '1992%';""", conn)
```

```python
pd.read_sql("""SELECT player_name, birthday
                FROM Player
                WHERE player_name LIKE 'Michael%' AND birthday LIKE '1992%';""", conn)
```
