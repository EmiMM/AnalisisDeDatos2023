```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 1. Leer archivo y explorar el dataset
df = pd.read_csv('GCB2022v27_MtCO2_flat.csv')
df
```

```python
df.set_index('Country', inplace=True)
df
```

```python
df.info()
```

```python
# 2. Obtener datos estadísticos de México

df_mx = df.loc['Mexico']
df_mx
```

```python
years_mx = df_mx['Year']

total_mx = df_mx['Total']
#years_mx, total_mx

plt.plot(years_mx, total_mx)
plt.title('Total Mexico')
```

```python
# datos globales

df_g = df.loc['Global']
df_g
```

```python
years_g = df_g['Year']

total_g = df_g['Total']
#years_g, total_g

plt.plot(years_g, total_g)
plt.title('Total Global')
```

```python
# 3. Comparar los valores globales con los de México

df_mx.compare(df_g)
```

```python
# 4. Gráfica de comparación de valores totales

plt.plot(years_mx, total_mx)
plt.plot(years_g, total_g)
plt.title('Total Mexico vs Total Global')
# plt.legend()
plt.show()
```
